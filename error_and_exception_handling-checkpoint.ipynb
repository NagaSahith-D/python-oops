### try, except & finally

```python
try: 
    f = open('testfile', 'w')
    f.write("write a test line")
except:
    print("All the exceptions")
finally:
    print("Doesnt matter, i always run")
```

```python
def ask_for_int():
    
    try:
        result = int(input("Please provide an number: "))
    except: 
        print("Looks like thats not a number")
    finally:
        print("End of try/except/finally")
```

```python
ask_for_int()
```

### using while for try,except & finally

```python
def ask_for_int():
    
    while True:
        try:
            result = int(input("Please provide an number: "))
        except: 
            print("Looks like thats not a number")
            continue
        else:
            print("Yes thank you")
            break
        finally:
            print("End of try/except/finally")
```

```python
ask_for_int()
```
