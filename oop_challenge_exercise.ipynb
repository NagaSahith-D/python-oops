```python
class Account():

    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, depo_amt):
        self.balance = self.balance + depo_amt
        print(f"Amount {depo_amt} added to the balance")

    def withdrawl(self, withd_amt):
        if self.balance >= withd_amt:
            self.balance = self.balance - withd_amt
            print("Withdrawl Accepted")
        else:
            print("Sorry, no enough funds")

    def __str__(self):
        return f"Owner : {self.owner} \nBalance : {self.balance}"
```
